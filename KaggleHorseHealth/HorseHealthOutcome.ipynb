```python
from sklearn.decomposition import PCA
from sklearn.model_selection import cross_val_score
```

```python
x = train_df.drop(['id','outcome'],axis=1)
```

```python
y = train_df.outcome
```

```python
pca = PCA(n_components=24)
```

```python
pca.fit_transform(x)
```

```python
pca.explained_variance_ratio_
```

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier()
```

```python
def cross_val(pca_range,estimator):
    for i in range(5,pca_range+1):
        pca = PCA(n_components=i)
        pca_x = pca.fit_transform(x)
        cvs = cross_val_score(estimator=estimator,X=pca_x,y=y,scoring='accuracy',verbose=2,cv=5)
        print(f'i={i}')
        print(cvs.mean())
```

```python
cross_val(24,DecisionTreeClassifier(max_depth=2))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
cross_val(24,RandomForestClassifier(max_depth=5))
```

```python
from sklearn.svm import SVC
```

```python
cross_val(24,SVC())
```

```python
cross_val(24,SVC(kernel='rbf')) ## best so far
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
cross_val(24,GaussianNB())
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
cross_val(24,LogisticRegression())
```

```python
import xgboost as xgb
from xgboost import XGBClassifier
```

```python
xgb_classifier = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=2,
    random_state=42
)
```

```python
cross_val(24,xgb_classifier)
```

```python
cross_val(24,hist_md)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
import tensorflow as tf
```

```python
model = tf.keras.models.Sequential()
model.add(tf.keras.Input(shape=(24,)))
model.add(tf.keras.layers.Dense(100,activation='relu'))
model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(200,activation='relu'))
#model.add(tf.keras.layers.Dropout(0.5))
model.add(tf.keras.layers.Dense(100,activation='relu'))
#model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(100,activation='relu'))
#model.add(tf.keras.layers.Dropout(0.2))
model.add(tf.keras.layers.Dense(3,activation='softmax'))

model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

model.fit(xtrain, ytrain, epochs=30, validation_split=0.3)
```

```python
model.evaluate(xtest,ytest)
```

```python
result = model.predict(test_df.drop(['id','outcome'],axis=1))
```

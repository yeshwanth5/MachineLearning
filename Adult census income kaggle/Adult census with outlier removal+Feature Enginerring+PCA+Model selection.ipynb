```python
from sklearn.preprocessing import StandardScaler
```

```python
ss = StandardScaler()
```

```python
df_x = pd.DataFrame(ss.fit_transform(df.drop('>50K',axis=1)))
```

```python
df_x.columns=df.drop('>50K',axis=1).columns
```

```python
df_x
```

```python
df_y = df['>50K']
df_y
```

```python
from sklearn.model_selection import cross_val_score
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
pca = PCA(n_components=12)
x_pca=pca.fit_transform(df_x)
print(pca.explained_variance_ratio_)
cvs=cross_val_score(estimator = LogisticRegression(),X=x_pca,y=df_y,cv=5,scoring='accuracy').mean()
cvs
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import PCA
from sklearn.ensemble import RandomForestClassifier
pca = PCA(n_components=12)
x_pca=pca.fit_transform(df_x)
print(pca.explained_variance_ratio_)
cvs=cross_val_score(estimator = RandomForestClassifier(n_estimators=100,random_state=42),X=x_pca,y=df_y,cv=5,scoring='accuracy').mean()
cvs
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import PCA
from sklearn.naive_bayes import GaussianNB
pca = PCA(n_components=12)
x_pca=pca.fit_transform(df_x)
print(pca.explained_variance_ratio_)
cvs=cross_val_score(estimator = GaussianNB(),X=x_pca,y=df_y,cv=5,scoring='accuracy').mean()
cvs
```

**As we can see the best model is RandomForestClassifier**

```python
def get_K(sample,n_clusters_limit,jumps):
    cv = CountVectorizer()
    vcuisines =cv.fit_transform(sample.cuisines)
    vcuisines = pd.DataFrame(vcuisines.toarray(), columns=cv.get_feature_names_out())
    sample.set_index(np.arange(len(sample)),inplace=True)
    sample = pd.concat([sample,vcuisines],axis=1)
    sample.drop('cuisines',axis=1,inplace=True)
    sample = pd.concat([sample,pd.get_dummies(sample.type,drop_first=True,dtype='int64')],axis=1)
    sample.drop('type',inplace=True,axis=1)
    sample.rate=mn.fit_transform(sample[["rate"]])
    sample.cost_for_two=mn.fit_transform(sample[["cost_for_two"]])
    
    inertia_values = []
    for num_clusters in range(1, n_clusters_limit,jumps):
        kmeans = KMeans(n_clusters=num_clusters, random_state=42)
        kmeans.fit(sample)
        inertia_values.append(kmeans.inertia_)
    
    # Plot the elbow curve
    plt.figure(figsize=(30,30))
    plt.plot( inertia_values, marker='o')
    plt.xlabel('Number of Clusters')
    plt.ylabel('Inertia')
    plt.title('Elbow Curve for KMeans')
    #plt.xticks(range(1, n_clusters_limit,jumps))
    plt.show()
```

```python
def form_clusters(sample,sample_copy):
    cv = CountVectorizer()
    vcuisines =cv.fit_transform(sample.cuisines)
    vcuisines = pd.DataFrame(vcuisines.toarray(), columns=cv.get_feature_names_out())
    sample.set_index(np.arange(len(sample)),inplace=True)
    sample_copy.set_index(np.arange(len(sample_copy)),inplace=True)
    sample = pd.concat([sample,vcuisines],axis=1)
    sample.drop('cuisines',axis=1,inplace=True)
    sample = pd.concat([sample,pd.get_dummies(sample.type,drop_first=True,dtype='int64')],axis=1)
    sample.drop('type',inplace=True,axis=1)
    sample.rate=mn.fit_transform(sample[["rate"]])
    sample.cost_for_two=mn.fit_transform(sample[["cost_for_two"]])
    kmeans = KMeans(n_clusters=20, random_state=30)
    kmeans.fit(sample)
    sample_copy = pd.concat([sample_copy,pd.DataFrame(kmeans.predict(sample))],axis=1)
    sample_copy.rename(columns={0:"Clusters"},inplace=True)
    return sample_copy
    
```

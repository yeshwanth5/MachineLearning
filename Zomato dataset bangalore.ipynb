**Cleaning rest_type**

```python
df.rest_type.unique().shape
```

```python
lessthan500=df.rest_type.value_counts()[df.rest_type.value_counts()<500]
```

```python
lessthan500.shape
```

```python
df[df.rest_type=='Club'].shape
```

```python
df.rest_type.value_counts()[df.rest_type.value_counts()<500].shape
```

```python
'Bar' in lessthan500
```

```python
def change_rest_type(value):
    if value in lessthan500:
        return "Others"
    return value

df.rest_type=df.rest_type.apply(change_rest_type)
```
